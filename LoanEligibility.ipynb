# Preprocessing

```python
df.isnull().mean() * 100
```

```python
# First, let's remove all the null values we have from the credit_history variable,
# as we saw earlier this is a very important variable for the model,
# and we can't say that we have the applicant's credit history when we don't, 
# in terms of business this can be very dangerous.
```

```python
df = df[df['Credit_History'].notnull()]
```

```python
#Here we will fill the null values of the variables Gender, Married, Dependents and Self_Employed with the mode

df['Gender']=df['Gender'].fillna(df['Gender'].mode()[0])
df['Married']=df['Married'].fillna(df['Married'].mode()[0])
df['Dependents']=df['Dependents'].fillna(df['Dependents'].mode()[0])
df['Self_Employed']=df['Self_Employed'].fillna(df['Self_Employed'].mode()[0])
```

```python
#Here we will fill the null values of the variables LoanAmount and Loan_Amount_Term with the Mean
df['LoanAmount']=df['LoanAmount'].fillna(df['LoanAmount'].mean())
df['Loan_Amount_Term']=df['Loan_Amount_Term'].fillna(df['Loan_Amount_Term'].mean())
```

```python
df.isnull().mean() * 100
```

```python
#we have succesfully handeled the missing values in our dataset
```

```python
x=df.drop(columns=['Loan_Status'])
y=df['Loan_Status']
```

```python
x,y
```

## Label Encoding

```python
from sklearn.preprocessing import LabelEncoder
lb_x=LabelEncoder()
lb_y=LabelEncoder()
```

```python
x['Gender']=lb_x.fit_transform(x['Gender'])
x['Married']=lb_x.fit_transform(x['Married'])
x['Education']=lb_x.fit_transform(x['Education'])
x['Self_Employed']=lb_x.fit_transform(x['Self_Employed'])
x['Property_Area']=lb_x.fit_transform(x['Property_Area'])
x['Credit_History']=lb_x.fit_transform(x['Credit_History'])
```

```python
y=lb_y.fit_transform(y)
```

```python
x['Dependents'].value_counts()
```

```python
x['Dependents']=x['Dependents'].replace('3+', 3).astype(int)
```

## Train_Test_Split

```python
from sklearn.model_selection import train_test_split

x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.20)
```

## LogisticRegression

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import r2_score,accuracy_score,classification_report,confusion_matrix

Reg=LogisticRegression(max_iter=5000)
Reg.fit(x_train,y_train)
y_pred=Reg.predict(x_test)
```

```python
accuracy_score(y_test, y_pred)
```

```python
print(classification_report(y_test, y_pred))
```

```python
cm = confusion_matrix(y_test, y_pred)
sns.heatmap(cm, annot=True, cmap='coolwarm') 
plt.xlabel('Predicted')
plt.title('Confusion Matrix')
plt.ylabel('True')
plt.show()
```
